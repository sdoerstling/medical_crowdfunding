```
from transformers import AutoModelForTokenClassification, AutoTokenizer
import torch

model = AutoModelForTokenClassification.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")

```

```
label_list = [
    "O",       # Outside of a named entity
    "B-Problem",  # Beginning of a miscellaneous entity right after another miscellaneous entity
    "I-Problem",  # Miscellaneous entity
    "B-Test",   # Beginning of a person's name right after another person's name
    "I-Test",   # Person's name
    "B-Treatment",   # Beginning of an organisation right after another organisation
    "I-Treatment",   # Organisation
    "B-Procedure",   # Beginning of a location right after another location
    "I-Procedure"    # Location
]

sequence = "My mom has breast cancer."

# Bit of a hack to get the tokens with the special tokens
tokens = tokenizer.tokenize(tokenizer.decode(tokenizer.encode(sequence)))
inputs = tokenizer.encode(sequence, return_tensors="pt")

```

```
inputs
```

```
outputs = model(inputs)[0]
predictions = torch.argmax(outputs, dim=2)
```

```

print([(token, label_list[prediction]) for token, prediction in zip(tokens, predictions[0].tolist())])
```

```
from transformers import BertTokenizer
from transformers import BertModel
import torch
tokenizer = BertTokenizer.from_pretrained('emilyalsentzer/Bio_ClinicalBERT')
model = BertModel.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
input_ids = torch.tensor(tokenizer.encode("My mom has breast cancer.")).unsqueeze(0)  
outputs = model(input_ids)
```
