```
from transformers import pipeline
```

```
classifier = pipeline("zero-shot-classification", model='emilyalsentzer/Bio_ClinicalBERT')
```

```
sequence = "brain tumor"
candidate_labels = ["musculoskeletal", "cardiovascular", 'cancer']
hypothesis_template = 'This text is about {}.'

classifier(sequence, candidate_labels, hypothesis_template=hypothesis_template, multi_class=True)
```
