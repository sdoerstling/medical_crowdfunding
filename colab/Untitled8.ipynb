```
from transformers import pipeline
from transformers import AutoTokenizer, AutoModel, BertForTokenClassification
tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
model = BertForTokenClassification.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
```

```
nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "My mom has breast cancer and needs an mri."

ner_results = nlp(example)
```
