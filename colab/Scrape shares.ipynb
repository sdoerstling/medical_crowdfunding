# Define scraping function

```
def get_shares(urls):

  '''
  input: list of urls
  returns: list of response status, list of shares, list of followers
  '''

  #define results container
  resp = []
  shares = []
  followers = []

  #loop through urls
  for url in urls:
    
    #check response status
    try:
      r = requests.get(url)
      status_code = r.status_code
    except:
      resp.append(np.nan)
      shares.append(np.nan)
      followers.append(np.nan)
      continue

    if status_code == 200:
      resp.append(status_code)
      wd.get(url)
      spans = wd.find_elements_by_class_name("text-stat-value")
      
      #try to get shares span value
      try:
        share = spans[4].text
        shares.append(share)
      except:
        shares.append(np.nan)
      
      #try to get followers span value
      try:
        follower = spans[5].text
        followers.append(follower)
      except:
        followers.append(np.nan)

    else:
      resp.append(status_code)
      shares.append(np.nan)
      followers.append(np.nan)

  return resp, shares, followers
```

# Import additional packages and urls to scrape

```
import pandas as pd
import numpy as np
```

```
feed = pd.read_csv('/content/drive/MyDrive/Crowdfunding/urls_for_share_scraping.csv')
urls = feed['url'].tolist()
```

```
urls = np.array_split(urls,10)
```

# Run scrape

```
resp = []
shares = []
followers = []

for url in urls[3]:
  resp_, shares_, followers_ = get_shares([url])
  resp.append(resp_)
  shares.append(shares_)
  followers.append(followers_)
```
