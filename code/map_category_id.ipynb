## Map of category IDs

```python
from bs4 import BeautifulSoup
import pandas as pd
import re
import json
import requests
```

```python
id_map = [
    {"category" : "medical", "url" : "https://www.gofundme.com/start/medical-fundraising"},
    {"category" : "memorial", "url" : "https://www.gofundme.com/start/memorial-fundraising"},
    {"category" : "emergency", "url" : "https://www.gofundme.com/start/emergency-fundraising"},
    {"category" : "education", "url" : "https://www.gofundme.com/start/education-fundraising"},
    {"category" : "animals", "url" : "https://www.gofundme.com/discover/animal-fundraiser"},
    {"category" : "business", "url" : "https://www.gofundme.com/discover/business-fundraiser"},
    {"category" : "community", "url" : "https://www.gofundme.com/discover/community-fundraiser"},
    {"category" : "competition", "url" : "https://www.gofundme.com/discover/competition-fundraiser"},
    {"category" : "creative", "url" : "https://www.gofundme.com/discover/creative-fundraiser"},
    {"category" : "event", "url" : "https://www.gofundme.com/discover/event-fundraiser"},
    {"category" : "faith", "url" : "https://www.gofundme.com/discover/faith-fundraiser"},
    {"category" : "family", "url" : "https://www.gofundme.com/discover/family-fundraiser"},
    {"category" : "newlywed", "url" : "https://www.gofundme.com/discover/newlywed-fundraiser"},
    {"category" : "sports", "url" : "https://www.gofundme.com/discover/sports-fundraiser"},
    {"category" : "travel", "url" : "https://www.gofundme.com/discover/travel-fundraiser"},
    {"category" : "volunteer", "url" : "https://www.gofundme.com/discover/volunteer-fundraiser"},
    {"category" : "wishes", "url" : "https://www.gofundme.com/discover/wishes-fundraiser"},
    {"category" : "other", "url" : "https://www.gofundme.com/discover/other-fundraiser"}
]
```

```python
id_df = pd.DataFrame(id_map)
```

```python
def getData(df):
    
    campaign_labels = []
    category_ids = []

    for i in range(0, len(df)):
        url = df['url'].values[i]
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'lxml')
        page_url = soup.find_all("a", class_="campaign-tile-img--contain")[0]['href']
        page_html = requests.get(page_url).text
        soup = BeautifulSoup(page_html, 'lxml')
        campaign_label = soup.find_all("a",{"class":"m-campaign-byline-type"})[0].text
        category_id = json.loads(re.findall(r'window\.initialState = ({.*?});', page_html)[0])['feed']['campaign']['category_id']
        campaign_labels.append(campaign_label)
        category_ids.append(category_id)
        
    return [campaign_labels, category_ids]
```

```python
data = getData(id_df)
```

```python
id_df['label'] = data[0]
id_df['category_id'] = data[1]
```
